```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("W03_data-UseC_-HR-Employee-Attrition.csv")

print(df.shape)  
df.head()
```

```python
# 기본 구성 확인하기
df.info()

df.describe()
```

```python
# null 값 확인 
print(df.isnull().sum())

for col in df.columns:
    if df[col].nunique() == 1:
        print("상수 컬럼:", col)
```

```python
# 타겟은 퇴사
print(df["Attrition"].value_counts())

# 막대그래프
plt.rc('font', family='AppleGothic')
df["Attrition"].value_counts().plot(kind="bar")
plt.title("Attrition 분포")
plt.xlabel("Attrition (No=잔류, Yes=이탈)")
plt.ylabel("개수")
plt.show()
```

```python
# 나이, 월급, 근속연수 히스토그램
for col in ["Age", "MonthlyIncome", "YearsAtCompany"]:
    df[col].hist(bins=30)
    plt.title(f"{col} 분포")
    plt.xlabel(col)
    plt.ylabel("빈도수")
    plt.show()
```

```python
# Attrition에 따른 월급 분포
df.boxplot(column="MonthlyIncome", by="Attrition")
plt.title("월급 - 퇴사 여부 비교")
plt.suptitle("")    
plt.xlabel("퇴사")
plt.ylabel("월급")
plt.show()

# Attrition에 따른 나이 분포
df.boxplot(column="Age", by="Attrition")
plt.title("나이 - 퇴사 여부 비교")
plt.suptitle("")
plt.xlabel("퇴사")
plt.ylabel("나이")
plt.show()
```

```python
# 부서별 퇴사율
ct = pd.crosstab(df["Department"], df["Attrition"], normalize="index")
print(ct)

ct.plot(kind="bar", stacked=True)
plt.title("부서별 퇴사율")
plt.ylabel("비율")
plt.show()

# 야근 여부와 퇴사율
ct2 = pd.crosstab(df["OverTime"], df["Attrition"], normalize="index")
print(ct2)

ct2.plot(kind="bar", stacked=True)
plt.title("야근 여부에 따른 퇴사율")
plt.ylabel("비율")
plt.show()
```

```python
# 직무 × 출장빈도 조합별 퇴사율
temp = df.copy()
temp["AttrBin"] = temp["Attrition"].map({"Yes":1,"No":0})
pivot = temp.pivot_table(index="JobRole", columns="BusinessTravel", values="AttrBin", aggfunc="mean")

plt.imshow(pivot, cmap="Reds", aspect="auto")
plt.xticks(range(len(pivot.columns)), pivot.columns)
plt.yticks(range(len(pivot.index)), pivot.index)
plt.colorbar(label="퇴사율")
plt.title("직무 × 출장빈도 퇴사율 Heatmap")
plt.show()
```

```python


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt



plt.rc('font', family='AppleGothic') 
plt.rcParams['axes.unicode_minus'] = False  

df = pd.read_csv("W03_data-UseC_-HR-Employee-Attrition.csv")
df["MonthlyIncome"] = pd.to_numeric(df["MonthlyIncome"], errors="coerce")

df["Attrition"] = df["Attrition"].map({"Yes": 1, "No": 0})


print("Attrition 고유값:", df["Attrition"].dropna().unique()[:10])
print("MonthlyIncome 결측:", df["MonthlyIncome"].isna().sum())
print("MonthlyIncome 유니크:", df["MonthlyIncome"].nunique())

# 소득 10분위(Decile) 만들기

mi_rank = df["MonthlyIncome"].rank(method="first")  # 동점값에 임의 순서 부여
df["IncomeDecile"] = pd.qcut(mi_rank, q=10, labels=False, duplicates="drop")

print("IncomeDecile 고유값:", sorted(df["IncomeDecile"].dropna().unique().tolist()))

# 소득 분위별 평균 퇴사율 계산 (NaN 제거)
tmp = df.dropna(subset=["IncomeDecile", "Attrition"]).copy()
attr_rate = tmp.groupby("IncomeDecile")["Attrition"].mean().sort_index()

print("\n소득 Decile별 퇴사율(비율):")
print(attr_rate)

# 6) 
ax = attr_rate.plot(kind="bar", figsize=(7,4))
plt.title("소득 Decile별 퇴사율")
plt.xlabel("소득 분위 (0=가장 낮음, 9=가장 높음)")
plt.ylabel("퇴사율")
plt.ylim(0, 1)

# 막대 레이블(%) 표시
for i, v in enumerate(attr_rate.values):
    plt.text(i, v + 0.02, f"{v*100:.1f}%", ha="center", va="bottom", fontsize=9)

plt.tight_layout()
plt.show()
```

```python
bins = [18, 29, 39, 49, 60]
labels = ["20대", "30대", "40대", "50대+"]
df["AgeGroup"] = pd.cut(df["Age"], bins=bins, labels=labels)

attr_by_age = df.groupby("AgeGroup")["Attrition"].mean()

attr_by_age.plot(kind="bar")
plt.title("연령대별 퇴사율")
plt.ylabel("퇴사율")
plt.show()
```

```python
attr_by_years = df.groupby("YearsAtCompany")["Attrition"].mean()

attr_by_years.plot(marker="o")
plt.title("근속연수별 퇴사율")
plt.xlabel("근속연수")
plt.ylabel("퇴사율")
plt.show()
```

```python
pivot = df.pivot_table(index="JobSatisfaction", columns="OverTime", values="Attrition", aggfunc="mean")

plt.imshow(pivot, cmap="Reds", aspect="auto")
plt.xticks(range(len(pivot.columns)), pivot.columns)
plt.yticks(range(len(pivot.index)), pivot.index)
plt.colorbar(label="퇴사율")
plt.title("직무만족도 × 야근 여부 → 퇴사율")
plt.show()
```
